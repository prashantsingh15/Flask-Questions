# Q3. What is App routing in Flask? Why do we use app routes?

- In Flask app routing refers to the process of defining URL routes within our web application. URL routes are the paths or URLs that users can access to interact with different parts of our application. By using app routes, we can  specify what code should be executed when a user accesses a specific URL, allowing you to control the behavior of our application.


- App routing is essential in Flask (and other web frameworks) because it allows you to structure our application, handle different user interactions, and provide content or data dynamically based on the URLs users visit. It enables you to create a logical flow for our web application and handle various use cases efficiently.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

- In Flask, the url_for() function is used for URL building. It generates a URL for the specified endpoint, taking into account any route parameters or query parameters that need to be included in the URL.

```python
from flask import Flask, url_for

app = Flask(__name__)


@app.route('/')
def index():
    return "This is the index page."

@app.route('/user/<username>')
def show_user_profile(username):
    return f"User: {username}"

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f"Post ID: {post_id}"


with app.test_request_context():
    print(url_for('index'))
    print(url_for('show_user_profile', username='prashant'))
    print(url_for('show_post', post_id=123))

if __name__ == '__main__':
    app.run()
```
